```python
import pandas as pd
import numpy as np
import joblib 

from sklearn.linear_model import LinearRegression
```

```python
X_train=pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\ML Project\Smartwatches project\Dataset\Processed\X_train.csv")
X_test=pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\ML Project\Smartwatches project\Dataset\Processed\X_test.csv")
y_train=pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\ML Project\Smartwatches project\Dataset\Processed\y_train.csv").squeeze()
y_test=pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\ML Project\Smartwatches project\Dataset\Processed\y_test.csv").squeeze()
```

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
model=LinearRegression()
```

```python
y_pred=model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
len(model.coef_)
```

```python
coef_df=pd.DataFrame({"feature":X_train.columns, "Coefficient": model.coef_})
coef_df= coef_df.sort_values(by='Coefficient', ascending=False)
print(coef_df.head())
print()
print(coef_df.tail())
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

```python
comparison=pd.DataFrame({"Actual Ratings":y_test, "Predicted Ratings": y_pred})
```
